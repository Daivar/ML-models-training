```python
from sklearn.datasets import load_iris 
from sklearn.tree import DecisionTreeClassifier 
 
iris = load_iris() 
# X = iris.data[:, 2:] # petal length and width 
X = iris.data
y = iris.target 
```

```python
iris.feature_names[2:]
```

```python
tree_clf = DecisionTreeClassifier(max_depth=100, random_state=42) 
tree_clf.fit(X, y)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y, tree_clf.predict(X)))
print(classification_report(y, tree_clf.predict(X)))
```

```python
from dtreeviz.trees import dtreeviz # remember to load the package

viz = dtreeviz(tree_clf, X, y,
                target_name="target",
                feature_names=iris.feature_names[2:],
                # feature_names=iris.feature_names,
                class_names=list(iris.target_names),
                scale=2.0)

viz
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(tree_clf, 
                  feature_names=iris.feature_names[2:],  
                   # feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from dtreeviz.trees import dtreeviz
iris = load_iris() 
X = iris.data[:, 1:]
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=5, random_state=142) 
tree_clf.fit(X, y)
print(confusion_matrix(y, tree_clf.predict(X)))
print(classification_report(y, tree_clf.predict(X)))
viz = dtreeviz(tree_clf, X, y,
                target_name="target",
                feature_names=iris.feature_names[1:],
                class_names=list(iris.target_names),
                scale=1.0)
viz
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import random

def print_paired(p1, p2):
  print("-----")
  for i in zip(p1, p2):
    print(f'{i[0]} -> {i[1]}')

# X = np.array([1, 2, 3, 4, 5])
X = np.arange(10)

# no noise
y = np.dot(X, 2) + 3

# with noise
y = np.dot(X, 2)
for i, y_val in enumerate(y):
  y[i] += random.randint(0,5)

print_paired(X, y)

reg = LinearRegression().fit(X.reshape(-1, 1), y) # reshape to get the data into the form: [[1], [2]]

plt.scatter(X, y, color='red', marker= '+', label= 'Training Data')
plt.plot(X, reg.predict(np.arange(X.shape[0]).reshape(-1, 1)), color='green', label='Linear Regression')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Fit')
plt.legend()

print("===== metrics =====")
print(reg.score(X.reshape(-1, 1), y)) # r^2
print(reg.coef_)
print(reg.intercept_)
print(f'f(x)={reg.coef_[0]}x+{reg.intercept_}')
```

```python
# adding data with noise

def print_paired(p1, p2):
  print("-----")
  for i in zip(p1, p2):
    print(f'{i[0]} -> {i[1]}')

X = np.arange(1000)
y = np.dot(X, 2)
for i, y_val in enumerate(y):
  y[i] += random.randint(0, i * 10 - i)

print_paired(X, y)
```

```python
plt.scatter(X, y, color='red', marker= '+', label= 'Training Data')
```

```python
import csv

f = open('generated_data.csv', 'w')
writer = csv.writer(f)

for i in zip(X, y):
  writer.writerow(i)

f.close()
```
